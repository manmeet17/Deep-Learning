```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
feat_cols=[tf.feature_column.numeric_column(key='x',shape=[1])]
feat_cols
```

```python
est=tf.estimator.LinearRegressor(feature_columns=feat_cols)
```

```python
from sklearn.model_selection import train_test_split
x_data=np.linspace(0.0,10.0,1000000)
noise=np.random.randn(len(x_data))
y_true=(0.5*x_data) + 5 + noise
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_data,y_true,test_size=0.3,random_state=101)
```

```python
x_train.shape
```

```python
inp_fun=tf.estimator.inputs.numpy_input_fn({'x':x_train},y_train,batch_size=8,num_epochs=None,shuffle=True)
```

```python
train_inp_fun=tf.estimator.inputs.numpy_input_fn({'x':x_train},y_train,batch_size=8,num_epochs=1000,shuffle=False)
```

```python
test_inp_fun=tf.estimator.inputs.numpy_input_fn({'x':x_test},y_test,batch_size=8,num_epochs=1000,shuffle=False)
```

```python
est.train(input_fn=inp_fun,steps=1000)
```

```python
train_metrics=est.evaluate(input_fn=train_inp_fun,steps=1000)
```

```python
test_metrics=est.evaluate(input_fn=test_inp_fun,steps=1000)
```

```python
print ("Training Metrics:")
print (train_metrics)
```

```python
print ("Eval Metrics:")
print (test_metrics)
```
